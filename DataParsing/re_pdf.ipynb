```python

import re
file = open("Air/4.TXT")
row_data = file.read()
expression = r'\d{2}\s*\w{3}\s*-\s*\d{2}\s*\w{3}[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*H\d{1,2}M'
result = re.findall(expression, row_data)
month = {
    'Nov': '11.18',
    'Dec': '12.18',
    'Jan': '01.19'
}
airplanes = {
    '100': '100',
    '310': '310',
    '313': '313',
    '318': '318',
    '319': '319',
    '320': '320',
    '321': '321',
    '32S': '32S',
    '330': '330',
    '332': '332',
    '333': '333',
    '340': '340',
    '343': '343',
    '345': '345',
    '346': '346',
    '380': '380',
    '388': '388',
    '717': '717',
    '733': '733',
    '734': '734',
    '735': '735',
    '736': '736',
    '737': '737',
    '738': '738',
    '739': '739',
    '73C': '73C',
    '73G': '73G',
    '73H': '73H',
    '73J': '73J',
    '73R': '73R',
    '73W': '73W',
    '744': '744',
    '747': '747',
    '74E': '74E',
    '74M': '74M',
    '752': '752',
    '753': '753',
    '757': '757',
    '75W': '75W',
    '762': '762',
    '763': '763',
    '764': '764',
    '767': '767',
    '76W': '76W',
    '772': '772',
    '773': '773',
    '777': '777',
    '77L': '77L',
    '77W': '77W',
    '787': '787',
    '788': '788',
    'A81': 'A81',
    'AB6': 'AB6',
    'AR8': 'AR8',
    'AT5': 'AT5',
    'AT7': 'AT7',
    'ATR': 'ATR',
    'BUS': 'BUS',
    'CR2': 'CR2',
    'CR7': 'CR7',
    'CR9': 'CR9',
    'CRJ': 'CRJ',
    'CRK': 'CRK',
    'DH4': 'DH4',
    'E70': 'E70',
    'E75': 'E75',
    'E90': 'E90',
    'E95': 'E95',
    'EM2': 'EM2',
    'EMJ': 'EMJ',
    'EQV': 'EQV',
    'ER3': 'ER3',
    'ER4': 'ER4',
    'ERJ': 'ERJ',
    'F50': 'F50',
    'F70': 'F70',
    'M11': 'M11',
    'M88': 'M88',
    'M90': 'M90',
    'S20': 'S20',
    'SF3': 'SF3',
    'SU9': 'SU9',
    'TRN': 'TRN'
}
with open('SKYTEAM.TXT', 'a') as f:
    i = 0
    for line in result:
        day_1 = re.findall('(\d{2})\s*\w{3}\s*-\s*\d{2}\s*\w{3}[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*M', line)
        month_1 = re.findall('\d{2}\s*(\w{3})\s*-\s*\d{2}\s*\w{3}[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*M', line)
        day_2 = re.findall('\d{2}\s*\w{3}\s*-\s*(\d{2})\s*\w{3}[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*M', line)
        month_2 = re.findall('\d{2}\s*\w{3}\s*-\s*\d{2}\s*(\w{3})[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*M', line)
        numbers = re.findall('\d{2}\s*\w{3}\s*-\s*\d{2}\s*\w{3}([0-9\s]*)\d{2}:\d{2}\d{2}:\d{2}[0-9A-Z\*]*M', line)
        time_1 = re.findall('\d{2}\s*\w{3}\s*-\s*\d{2}\s*\w{3}[0-9\s]*(\d{2}:\d{2})\d{2}:\d{2}[0-9A-Z\*]*M', line)
        time_2 = re.findall('\d{2}\s*\w{3}\s*-\s*\d{2}\s*\w{3}[0-9\s]*\d{2}:\d{2}(\d{2}:\d{2})[0-9A-Z\*]*M', line)
        flight = re.findall('\d{2}\s*\w{3}\s*-\s*\d{2}\s*\w{3}[0-9\s]*\d{2}:\d{2}\d{2}:\d{2}([0-9A-Z\*]*M)', line)
        if month_1[0] in month:
            month_1_n = month[month_1[0]]
        if month_2[0] in month:
            month_2_n = month[month_2[0]]
        date_1 = str(day_1[0] + '.' + month_1_n)
        date_2 = str(day_2[0] + '.' + month_2_n)
        numbers = numbers[0].replace(' ', '')
        flight = flight[0]
        flight = flight.replace('*', '')
        if flight[6:9] in airplanes and flight[9] != 'H':
            flight_number = flight[0:6]
            flight_plane = flight[6:9]
            flight_duration = flight[9:]
            flight_duration = flight_duration.replace('H', ':')
            flight_duration = flight_duration.replace('M', '')
            if len(flight_duration) != 5 and flight_duration[1] == ':':
                flight_duration = '0' + flight_duration
            if len(flight_duration) != 5 and flight_duration[2] == ':':
                flight_duration = flight_duration[0:3] + '0' + flight_duration[3]
        elif flight[5:8] in airplanes and flight[8] != 'H':
            flight_number = flight[0:5]
            flight_plane = flight[5:8]
            flight_duration = flight[8:]
            flight_duration = flight_duration.replace('H', ':')
            flight_duration = flight_duration.replace('M', '')
            if len(flight_duration) != 5 and flight_duration[1] == ':':
                flight_duration = '0' + flight_duration
            if len(flight_duration) != 5 and flight_duration[2] == ':':
                flight_duration = flight_duration[0:3] + '0' + flight_duration[3]
        elif flight[4:7] in airplanes:
            flight_number = flight[0:4]
            flight_plane = flight[4:7]
            flight_duration = flight[7:]
            flight_duration = flight_duration.replace('H', ':')
            flight_duration = flight_duration.replace('M', '')
            if len(flight_duration) != 5 and flight_duration[1] == ':':
                flight_duration = '0' + flight_duration
            if len(flight_duration) != 5 and flight_duration[2] == ':':
                flight_duration = flight_duration[0:3] + '0' + flight_duration[3]
        else:
            flight_number = flight[0:3]
            flight_plane = flight[3:6]
            flight_duration = flight[6:]
            flight_duration = flight_duration.replace('H', ':')
            flight_duration = flight_duration.replace('M', '')
            if len(flight_duration) != 5 and flight_duration[1] == ':':
                flight_duration = '0' + flight_duration
            if len(flight_duration) != 5 and flight_duration[2] == ':':
                flight_duration = flight_duration[0:3] + '0' + flight_duration[3]
        res = date_1 + ';' + date_2 + ';' + numbers + ';' + time_1[0] + ';' + time_2[0] + ';' + flight_number + ';' + flight_plane + ';' + flight_duration
        i += 1
#         print(res, i)
        f.write(str(res)  + "\n")
```
