```python
from heapq import nlargest

suspiciousValue = 3 # Сколько совпадающих полетов, чтобы считать подозрительным
topN = 2 # За сколькими людьми "следует" человек (когда считаем кол-во совпадений, сколько наибольших значений будет учитываться при анализе)

#########################################

def writeMatrix(arr):
    print()
    for i in arr:
        print(i)
    print()
    pass

text = open('INPUT.TXT').read()
text = text.split('\n')
result = {}
for el in text:
    line = el.split(";")
    # Если в словаре нет текущего пассажира, то добавляем его
    if line[0] not in result:
        result[line[0]] = set([])
    # Если у текущего пассажира нет текущего рейса, то добавляем его
    if line[1] not in result[line[0]]:
        result[line[0]].add(line[1])

# Ищем у кого меньше двух полетов
toremove = set([])
for el in result:
    if len(result[el]) < 2:
        toremove.add(el)

# Удаляем у кого меньше двух полетов
for el in toremove:
    print('Удаляем ', el, result[el])
    result.pop(el, None)
    
clientIdList = list(result.keys())
        
matches = []
for i in result:
    row = []
    for j in result:
        row += [0]
    matches += [row]

# Составляем матрицу совпадений, у кого с кем сколько одинаковых полетов
ci = 0
for i in result:
    cj = 0
    for j in result:
        if i != j:
#            print(i, ':', result[i], '&', j, ':', result[j])
            matches[ci][cj] = len(result[i].intersection(result[j]))
        cj += 1
    ci += 1    
writeMatrix(matches)

suspiciousPairs = []
# Смотрим у каждого пассажира кол-во совпадений с другими пассажирами и берем оттуда topN максимальных значений, если они >=suspiciousValue, то помечаем его, как подозрительного
for i in range(len(matches)):
    sortedIndexes = sorted(range(len(matches[i])), key=lambda j:matches[i][j], reverse=True)
    for j in range(topN):
#        print(matches[i][sortedIndexes[j]])
        if matches[i][sortedIndexes[j]] >= suspiciousValue:
            print('Very suspicious person with id', clientIdList[i], 'he might be following person', clientIdList[sortedIndexes[j]], '-_-')
            pair = [clientIdList[i], clientIdList[sortedIndexes[j]]]
            if pair not in suspiciousPairs and pair[::-1] not in suspiciousPairs:
                suspiciousPairs += [pair]
        else:
            print('Person with id', clientIdList[i], 'is not following person', clientIdList[sortedIndexes[j]])
    print()

print(suspiciousPairs)
print(result)
file = open('OUTPUT.TXT', 'w')
file.write(' '.join(result))
file.close()
```
